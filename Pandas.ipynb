```python
import numpy as np

# Нормализация матрицы (главная диагональ обнуляется, остальные элементы делятся на диагональный элемент)
def normalize_matrix(A):
    n = A.shape[0]
    A_norm = np.zeros_like(A)
    
    for i in range(n):
        diag = A[i, i]
        if diag == 0:
            raise ValueError(f"Элемент диагонали равен нулю в строке {i+1}, нормализация невозможна.")
        A_norm[i] = A[i] / diag
        A_norm[i, i] = 0  # Диагональные элементы обнуляем
    
    return A_norm

# Проверка условий сходимости
def check_convergence_conditions_normalized(A):
    # Нормализация матрицы
    A_norm = normalize_matrix(A)
    
    # Условие 1: alpha = max(sum(|a_ij_norm|, j != i)) для строк
    alpha = max(np.sum(np.abs(A_norm), axis=1))  # Сумма по строкам, исключая диагональные элементы (они уже равны 0)
    
    # Условие 2: beta = max(sum(|a_ij_norm|, i != j)) для столбцов
    beta = max(np.sum(np.abs(A_norm), axis=0))  # Сумма по столбцам, исключая диагональные элементы (они уже равны 0)
    
    # Условие 3: gamma = sum((a_ij_norm)^2) — сумма квадратов всех нормализованных элементов
    gamma = np.sum(A_norm**2)  # Сумма квадратов всех элементов нормализованной матрицы (диагональ уже нули)
    
    return alpha, beta, gamma

# Метод простой итерации для решения СЛАУ
def simple_iteration(A, b, epsilon, max_iter=1000, x0=None):
    n = A.shape[0]
    
    # Проверка условий сходимости
    alpha, beta, gamma = check_convergence_conditions_normalized(A)
    print(f"Проверка условий сходимости:")
    print(f"alpha = {alpha} (должно быть < 1)")
    print(f"beta = {beta} (должно быть < 1)")
    print(f"gamma = {gamma} (должно быть < 1)")
    
    if alpha >= 1 or beta >= 1:
        print("Предупреждение: Условия сходимости могут не выполняться!")
    
    # Инициализация начального приближения
    if x0 is None:
        x = np.zeros(n)  # Если начальное приближение не задано, используем вектор нулей
    else:
        x = np.array(x0, dtype=float)

    # Итерационный процесс
    A_norm = normalize_matrix(A)
    b_norm = b / np.diag(A)
    
    print("\nНачало итераций:")
    for k in range(1, max_iter + 1):
        x_new = np.copy(b_norm)
        for i in range(n):
            x_new[i] -= np.dot(A_norm[i], x)  # Формула итерации

        # Рассчитываем максимальную погрешность
        error = np.max(np.abs(x_new - x))

        # Вывод промежуточных результатов
        print(f"Итерация {k}: x = {x_new}, Погрешность = {error}")

        # Проверка на сходимость
        if error <= epsilon:
            print(f"\nПроцесс сошелся на итерации {k}.")
            return x_new
        
        x = x_new  # Обновление вектора для следующей итерации

    print(f"\nМетод не сошелся за {max_iter} итераций.")
    return x

# Ввод матриц с диагональным преоблоданием, свободных членов, и эпсилон
A = np.array([[5, 0.2, -0.7],
              [1, 3.8, -1.5],
              [1, -1, 7.3]], dtype=float)

b = np.array([1, -1, 0], dtype=float)
epsilon = 1e-5


x0 = [0, 0, 0]  # Начальное приближение

# Запуск метода простой итерации
solution = simple_iteration(A, b, epsilon, x0=x0)
print(f"\nРешение при эпсилон {epsilon}: {solution}")
```
